```python
import dask.dataframe as dd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Number of points in plot
num_points = 5000
```

```python
used_cars_hdf = 'used_cars_preprocessed.hdf'

# Read number of rows from preprocessed data
data = dd.read_hdf(used_cars_hdf, '/data-*')
data = data.head(num_points)
```

```python
from sklearn.linear_model import LinearRegression

# Get points from dataset
X = pd.DataFrame(data['mileage'])
y = pd.DataFrame(data['price'])

# Draw the points
plt.figure(figsize=(12, 8))
plt.scatter(X, y, alpha=0.3)

# Fit the linear regression
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

# Draw the linear regression line
plt.plot(X, y_pred, color='red', linewidth=4)
plt.title(f'Mileage vs. Price for {num_points} cars')
plt.xlabel('Mileage in miles')
plt.ylabel('Price in USD')
plt.xlim(0, 200000)
plt.ylim(0, 50000)
plt.show()
```
